Filter out Ho Chi Minh City

Time-based Grouping

```python
df_weather_Ho_Chi_Minh_City['year'] = df_weather_Ho_Chi_Minh_City['date'].dt.year
df_weather_Ho_Chi_Minh_City['month'] = df_weather_Ho_Chi_Minh_City['date'].dt.month
df_weather_Ho_Chi_Minh_City['day'] = df_weather_Ho_Chi_Minh_City['date'].dt.day
```

Create 30-Minute Time Bins (Matching Weather and Traffic Time Intervals

```python
# Round weather data to nearest 30-minute interval for easier comparison
df_weather_Ho_Chi_Minh_City['date_30min'] = df_weather_Ho_Chi_Minh_City['date'].dt.round('30T')

# Round traffic data to nearest 30-minute interval
df_segment_status['updated_at_30min'] = df_segment_status['updated_at'].dt.round('30T')
```

Merge Weather Data with Traffic Data on Time

```python
# Remove timezone from 'updated_at_30min' in df_segment_status
df_segment_status['updated_at_30min'] = df_segment_status['updated_at_30min'].dt.tz_localize(None)

# Merge weather and traffic data on 30-minute intervals
df_traffic_weather = pd.merge(df_segment_status, df_weather_Ho_Chi_Minh_City, left_on='updated_at_30min', right_on='date_30min', how='left')
```

Combine Segment and Node Information: Next, you can add geographic information by merging with df_segment and df_node

```python
# Merge traffic weather data with segment information
df_traffic_weather_segment = pd.merge(df_traffic_weather, df_segment, left_on='segment_id', right_on='_id', how='left')

# Merge segment information with node information to get geographical data
df_traffic_weather_segment_node = pd.merge(df_traffic_weather_segment, df_node, left_on='s_node_id', right_on='_id', how='left', suffixes=('_start', '_end'))
```

```python
print(df_traffic_weather_segment_node.tail())
```

 filter out rows with missing weather data

```python
# Filter out rows where weather data is missing
df_traffic_weather_segment_node_clean = df_traffic_weather_segment_node.dropna(subset=['province', 'max', 'min', 'wind', 'rain', 'humidi'])

# Check the cleaned dataframe
print(df_traffic_weather_segment_node_clean.tail())
```
